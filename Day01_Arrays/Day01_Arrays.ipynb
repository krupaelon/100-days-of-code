```python
#Sum of Array
arr=[1,2,5,6,8,0]
total=0
for i in arr:
  total+=i
print(total)
```

```python
#avg of array
arr=[1,2,5,6,8,0]
n=len(arr)
total=0
for i in  arr:
  total+=i
avg=total/n
print(avg)
```

```python
#EVEN ODD
arr = [1,2,3,4,5,6]
even,odd=0,0
for num in arr:
  if num % 2 == 0:
    even+=1
  else:
    odd +=1
print("Even numbers:",even)
print("Odd numbers:",odd)
```

```python
#Linear search
arr=[10,20,30,40,50]
target=50
found=False
for i in range(len(arr)):
  if arr[i]==target:
    print("Found at index ",i)
    found= True
    break
if not found:
  print("Not found")
```

```python
#reversing array without built in functions
arr=[10,20,30,40,50]
for i in range(len(arr)-1,-1,-1):
    print(arr[i])
```

```python
#find largets and smallest
arr=[2,8,4,0,5,1,3]
largest=arr[0]
smallest=arr[0]
for i in range(1,len(arr)):
  if arr[i]>largest:
    largest=arr[i]
  elif arr[i]<smallest:
    smallest=arr[i]
print("Largest ",largest)
print("smallest",smallest)
```

```python
#cheak if array sorted
arr=[4,7,2,5,1,0]
h1=sorted(arr)
if arr==h1:
  print("array is sorted")
else:
  print("not sorted")
```

```python
arr = [1,2,5,3,6]

flag = True

for i in range(len(arr)-1):
    if arr[i]>=arr[i+1]:
        flag = False
        break

if flag:
    print("Sorted")
else:
    print("Not Sorted")
```

```python
#find Duplicate
arr = [1,2,1,3,2,1,5]

seen = set()

for num in arr:

    if num in seen:
        print("Duplicate =", num)
        break

    seen.add(num)
```
